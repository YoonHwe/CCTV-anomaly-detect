```python
#두 점 사이의 각도 계산
import math
def getDegree(center_X, center_Y, k_X, k_Y):
  rad = math.atan2(center_Y - k_Y, center_X - k_X)
  PI = math.pi
  degree = (rad*180) / PI
  return degree
```
